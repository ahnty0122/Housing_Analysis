## 주택거래 매매, 전월세 데이터 merge
##### 같은 건물에 대해 매매와 전월세가 어떤 차이가 있는지 확인

```python
import pandas as pd
```

```python
# 주택거래_매매 데이터 불러오기
df1 = pd.read_csv('./data/주택거래_매매.csv')
df1
```

```python
# 주택거래_전월세 데이터 불러오기
df2 = pd.read_csv('./data/주택거래_전월세.csv', low_memory=False)
df2
```

```python
# 데이터들 열 확인
print(df1.columns)
print(df2.columns)
```

```python
# 아파트 매매, 전월세 데이터만 불러오기
df1_apart = df1[df1['TYPE']=='아파트']
df1_apart
```

```python
# 아파트 전세 데이터 불러오기
df2_apart = df2[(df2['TYPE']=='아파트') & (df2['전월세구분']=='전세')]
df2_apart
```

```python
merge_df = pd.merge(df1_apart, df2_apart, how = 'inner', on = ['지역명', '본번', '부번', '단지명', '전용면적(㎡)', '층', '건축년도', 'TYPE', '건물명', '도로조건', '행정동코드', '시군구', '읍면동명', '지역코드', '동리명'])
merge_df
```

```python
# 합쳐진 데이터의 열 목록 확인
merge_df.columns
```

```python
# 합쳐진 데이터에서 필요없는 열 제거
del merge_df['대지권면적(㎡)'], merge_df['연면적(㎡)'], merge_df['주택유형'], merge_df['도로조건']
merge_df
```

```python
# 열 목록 다시 확인
merge_df.columns
```

```python
merge_df.to_csv('./data/매매_전월세_비교.csv', index = False, encoding = 'utf-8')
```
