23.세종시 도로명주소 건물 geojson 파일에서 코드를 코드명으로 annotation

```python
import pandas as pd
```

```python
code_definition = pd.read_csv('건물용도코드.csv')
code_definition
```

```python
import geopandas as gpd

df = gpd.read_file('23.세종시_도로명주소_건물.geojson', encoding='cp949')
df
```

```python
# 23번 파일을 pandas dataframe 으로 변경
df2 = pd.DataFrame(df)
df2
```

```python
# 다른 데이터프레임과 merge 위해 type int로 변경해 type 맞춰줌
df2['BDTYP_CD'] = df2['BDTYP_CD'].astype('int')
```

```python
# 건물용도코드로 annotation
code_merge = pd.merge(df2, code_definition, left_on = 'BDTYP_CD', right_on = '코드', how = 'left')
code_merge
```

```python
# 건물구분코드로 annotation
build_se = []
for code in code_merge['BULD_SE_CD']:
    if code == '0':
        build_se.append('지상')
    elif code == '1':
        build_se.append('지하')
    elif code == '2':
        build_se.append('공중')
    else:
        build_se.append('')
code_merge['build_se'] = build_se
```

```python
code_merge
```

```python
code_merge.columns
```

```python
# annotation 결과 파일 열 이름 변경
code_merge.columns = ['BDTYP_CD', 'BULD_NM', 'BULD_NM_DC', 'BULD_SE_CD', 'BUL_MAN_NO',
       'EMD_CD', 'GRO_FLO_CO', 'LNBR_MNNM', 'LNBR_SLNO', 'UND_FLO_CO',
       'geometry', '코드', 'BDTYP_CD_NM', 'BULD_SE_NM']
```

```python
del code_merge['코드']
```

```python
code_merge
```

```python
code_merge.to_csv('23_세종시도로명주소건물.csv', index = False, encoding = 'utf-8') # 23번 파일 annotation 결과 csv 파일로 저장
```
