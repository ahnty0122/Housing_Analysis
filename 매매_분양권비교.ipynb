## 주택 매매, 분양권 거래 데이터 join
##### 같은 주택 유형에 대해 매매와 분양권 거래가 어떤 차이가 있는지 확인

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/주택거래_매매.csv', low_memory = False)
df1
```

```python
df2 = pd.read_csv('./data/분양권_거래.csv', low_memory = False)
df2
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# 매매 데이터에서 아파트만 불러옴 
df1_apart = df1[df1['TYPE'] == '아파트']
```

```python
# 같은 값 가지는 데이터를 기준으로 데이터 합쳐 같은 건물에 대한 분양권, 매매 데이터 비교
merge_df = pd.merge(df1_apart, df2, how = 'inner', on = ['단지명', '지역코드', '전용면적(㎡)', '지역명', '층', '행정동코드', '시군구', '읍면동명', '동리명'])
merge_df
```

```python
# 열 목록 확인
merge_df.columns
```

```python
# 필요 없는 열 제거
del merge_df['대지권면적(㎡)'], merge_df['주택유형'], merge_df['도로조건'], merge_df['대지면적(㎡)'], merge_df['해제사유발생일_x'], merge_df['해제사유발생일_y']
```

```python
merge_df
```

```python
merge_df.to_csv('./data/매매_분양권_비교.csv', index = False, encoding = 'utf-8')
```
