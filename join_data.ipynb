## Data JOIN
##### 아파트, 연립다세대, 단독다가구, 오피스텔 데이터 JOIN

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/7.세종시_아파트(전월세)_실거래가.csv')
df1
```

```python
df2 = pd.read_csv('./data/8.세종시_연립다세대(전월세)_실거래가.csv')
df2
```

```python
df3 = pd.read_csv('./data/9.세종시_단독다가구(전월세)_실거래가.csv')
df3
```

```python
df4 = pd.read_csv('./data/10.세종시_오피스텔(전월세)_실거래가.csv')
df4
```

```python
# 3(아파트), 4(연립다세대), 5(단독다가구), 6(오피스텔) 데이터 type 지정해주는 열 추가
df1['TYPE'] = '아파트'
df2['TYPE'] = '연립'
df3['TYPE'] = '단독'
df4['TYPE'] = '오피스텔'
```

```python
# 아파트, 연립다세대, 단독다가구, 오피스텔 데이터 합치면서 없는 값은 NaN 값으로 들어감
join_df = pd.concat([df1, df2, df3, df4], sort = False)
```

```python
# 시군구 열 이름을 지역명으로 변경
join_df.rename(columns = {'시군구': '지역명'}, inplace = True)
join_df
```

## '지역명'을 key로 설정해 아파트, 연립, 단독, 오피스텔 데이터와 지역코드 데이터 join

```python
code = pd.read_csv('./data/지역코드_mix.csv')
code
```

```python
final_df = pd.merge(join_df, code, on = '지역명')
final_df
```

```python
final_df.to_csv('./data/주택거래_전월세.csv', index = False, encoding = 'utf-8')
```
