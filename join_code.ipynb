## 지역코드와 데이터 조인
##### 상업업무용 거래, 토지 거래, 분양권 거래 데이터

```python
import pandas as pd
```

```python
# 기존 데이터 불러오기
df1 = pd.read_csv('./data/11.세종시_상업업무용(매매)_실거래가.csv')
df1
```

```python
# 시군구 열 이름을 지역명으로 변경
df1.rename(columns = {'시군구': '지역명'}, inplace = True)
```

```python
# join할 지역코드 데이터 불러오기
df2 = pd.read_csv('./data/지역코드_mix.csv')
df2
```

```python
join_df = pd.merge(df1, df2, on = '지역명')
join_df
```

```python
join_df.to_csv('./data/상업업무용_거래.csv', index = False, encoding = 'utf-8')
```
