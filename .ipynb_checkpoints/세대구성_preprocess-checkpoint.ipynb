## 세종시 지역별 세대원수별 세대수 데이터 전처리
##### 읍면동 열 공백 제거
##### 데이터프레임 행, 열 형태 변경

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/28.세종시_지역별_세대원수별_세대수.csv')
df
```

```python
df['읍면동'] = df['읍면동'].str.replace(" ","")
df['읍면동']
```

```python
# 세대원수 기준 변경
df['2인-3인'] = df['2인'] + df['3인']
df['4인-5인'] = df['4인'] + df['5인']
df['6인 이상'] = df['6인'] + df['7인'] + df['8인'] + df['9인'] + df['10인이상']
del df['2인'], df['3인'], df['4인'], df['5인'], df['6인'], df['7인'], df['8인'], df['9인'], df['10인이상']
```

```python
df
```

```python
# 세종시 전체 세대원수별 세대수 생성
a = df['계'].sum()
b = df['1인'].sum()
c = df['2인-3인'].sum()
d = df['4인-5인'].sum()
e = df['6인 이상'].sum()
# 세종특별자치시 행 추가
df.loc['19'] = ['세종특별자치시', a, b, c, d, e]
df
```

```python
df.iloc[19][1:]
```

```python
# 동별로 복제될 세종특별자치시 인구수 (5행)
new = pd.DataFrame(df.iloc[19][1:])
# index column 열 생성
new = new.reset_index()
new
```

```python
# 열 이름 변경
new.columns = ['구분', '세종시전체']
new
```

```python
df['읍면동']
```

```python
# 동 개수만큼 세대원수별 세대수 데이터 추가 (20 * 5 = 100)
new2 = new
for i in range(19):
    new = new.append(new2)
new
```

```python
# 동 개수 list 생성
dong_list = df['읍면동'].unique()
dong_list
```

```python
# 동 list도 데이터프레임 행 개수 (20*5=100) 로 만들어줌
new_dong_list = []
for i in dong_list:
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
new_dong_list
```

```python
# 데이터프레임에 읍면동 열 추가 후 동 이름 할당
new['읍면동'] = new_dong_list
new
```

```python
# 동 별 세대수 list 생성
sedae = []
for i in range(20):
    sedae.append(list(df.iloc[i][1:]))
sedae
```

```python
# 이제 세대수를 데이터프레임 개수만큼 할당하기 위해 세대수 2차원 list를 1차원 list로 변경 (5 * 21 = 105)
sedae2 = sum(sedae, [])
len(sedae2)
```

```python
# 동별세대수 열 생성
new['동별세대수'] = sedae2
new
```

```python
# 최종 데이터프레임 열 순서 변경
new = new[['읍면동', '동별세대수', '세종시전체', '구분']]
new
```

```python
new.to_csv('./data/28_세대구성.csv', index = False, encoding = 'utf-8')
```
