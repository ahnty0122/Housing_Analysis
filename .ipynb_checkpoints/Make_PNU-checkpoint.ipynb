### 표제부 PNU 추가
#### (PNU: 00000-00000-0-0000-0000 총 19자리로 만들어주기)

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/1.세종시_전유부.csv')
df1
```

```python
# 네 자리 숫자로 변경하기 위해 str type으로 바꿔줌
df1['시군구코드'] = df1['시군구코드'].astype(str)
df1['법정동코드'] = df1['법정동코드'].astype(str)
df1['대지구분코드'] = df1['대지구분코드'].astype(str)
df1['번'] = df1['번'].astype(str)
df1['지'] = df1['지'].astype(str)
```

```python
# '번'과 '지' 데이터를 네 자리 숫자로 변경
df1['번'] = df1['번'].str.zfill(4)
df1['지'] = df1['지'].str.zfill(4)
```

```python
df1
```

```python
# 시군구코드-법정동코드-대지구분코드-번-지(00000-00000-0000-0000) 연결해 PNU 생성
df1['PNU'] = df1['시군구코드'] + df1['법정동코드'] + df1['대지구분코드'] + df1['번'] + df1['지']
```

```python
df1
```

```python
# PNU 생성한 데이터 저장
df1.to_csv('./data/1_전유부_PNU.csv', index = False, encoding = 'utf-8')
```
