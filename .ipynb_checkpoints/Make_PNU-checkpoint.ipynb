### 표제부 PNU 추가
#### (PNU: 00000-00000-0-0000-0000 총 19자리로 만들어주기)

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./2.세종시_표제부.csv', encoding = 'cp949')
df1
```

```python
df1['시군구코드'] = df1['시군구코드'].astype(str)
df1['법정동코드'] = df1['법정동코드'].astype(str)
df1['대지구분코드'] = df1['대지구분코드'].astype(str)
df1['번'] = df1['번'].astype(str)
df1['지'] = df1['지'].astype(str)
```

```python
# 네 자리 숫자로 변경
df1['번'] = df1['번'].str.zfill(4)
df1['지'] = df1['지'].str.zfill(4)
```

```python
df1
```

```python
df1['PNU'] = df1['시군구코드'] + df1['법정동코드'] + df1['대지구분코드'] + df1['번'] + df1['지']
```

```python
df1
```

```python
df1.to_csv('2_표제부_PNU.csv', index = False, encoding = 'utf-8')
```
