### 23.세종시 도로명주소 건물 geojson 파일에서 코드를 코드명으로 annotation

```python
import pandas as pd
```

```python
# 35번 코드정의서에 있던 건물용도 코드의 코드, 코드명 파일 불러옴
code_definition = pd.read_csv('./data/건물용도코드.csv')
code_definition
```

```python
import geopandas as gpd
# geojson을 데이터프레임 형식으로 가져옴
df = gpd.read_file('./data/23.세종시_도로명주소_건물.geojson')
df
```

```python
# 23번 파일을 pandas dataframe 으로 변경
df2 = pd.DataFrame(df)
df2
```

```python
# 건물용도코드명 데이터프레임과 merge하기 위해 type을 int로 맞춰줌
df2['BDTYP_CD'] = df2['BDTYP_CD'].astype('int')
```

```python
# 건물용도코드로 annotation
# df2(세종시 도로명주소 건물 데이터)와 code_definition(건물용도코드 데이터)를 코드 열을 기준으로 merge
code_merge = pd.merge(df2, code_definition, left_on = 'BDTYP_CD', right_on = '코드', how = 'left')
code_merge
```

```python
# 건물구분코드로 annotation
# 건물구분코드는 0일 때 지상, 1일 때 지하, 2일 때 공중
# build_se 라는 배열 만들어 건물구분문자열 저장
build_se = []
for code in code_merge['BULD_SE_CD']:
    if code == '0':
        build_se.append('지상')
    elif code == '1':
        build_se.append('지하')
    elif code == '2':
        build_se.append('공중')
    else:
        build_se.append('')
# build_se열 생성해 건물구분문자열 할당
code_merge['build_se'] = build_se
```

```python
code_merge
```

```python
# annotation후 데이터프레임 열 이름 변경 위해 열 이름 확인
code_merge.columns
```

```python
# annotation 결과 파일 열 이름 변경
code_merge.columns = ['BDTYP_CD', 'BULD_NM', 'BULD_NM_DC', 'BULD_SE_CD', 'BUL_MAN_NO',
       'EMD_CD', 'GRO_FLO_CO', 'LNBR_MNNM', 'LNBR_SLNO', 'UND_FLO_CO',
       'geometry', '코드', 'BDTYP_CD_NM', 'BULD_SE_NM']
```

```python
# 코드 열 중복되므로 코드 열 삭제
del code_merge['코드']
```

```python
code_merge
```

```python
# 최종 데이터프레임 저장
code_merge.to_csv('./data/23_세종시도로명주소건물.csv', index = False, encoding = 'utf-8') # 23번 파일 annotation 결과 csv 파일로 저장
```
