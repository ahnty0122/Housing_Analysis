### 3번 파일에 지역코드 데이터 추가하기

```python
import pandas as pd
```

```python
# 지역코드데이터(외부데이터) 불러오기
jscode = pd.read_csv('./data/jscode.csv')
jscode
```

```python
# 위의 데이터프레임에서 동리, 지역코드 열만 가져오기
jscode.loc[:,['동리', '지역코드']]
```

```python
# 지역명이 시군구-동리-읍면동으로 분리된 세종시 아파트 실거래가 데이터 불러오기
df = pd.read_csv('./3.아파트_split.csv')
df
```

```python
# df(아파트 실거래가 데이터)와 jscode(지역코드) 데이터를 '동리'열을 기준으로 merge
final = pd.merge(df, jscode.loc[:,['동리', '지역코드']], on = '동리')
final
```

```python
# 지역코드(jscode) 추가된 데이터 저장
final.to_csv('./data/3.아파트_join_jscode.csv', index = False, encoding = 'utf-8')
```
