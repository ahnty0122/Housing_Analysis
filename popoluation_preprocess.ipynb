## 인구현황 데이터 전처리

```python
import pandas as pd
```

```python
# 세종시 연령별 인구현황 중 2020년 12월 데이터 불러오기
# 남/녀 구분 없이 다섯개 기준의 연령대 데이터로 전처리한 인구현황 데이터
f = pd.read_csv('./data/202012인구.csv')
f
```

```python
f.iloc[0]
```

```python
# 동별로 복제될 세종특별자치시 인구수 (5행)
new = pd.DataFrame(f.iloc[0][1:])
# index column 열 생성
new = new.reset_index()
# 열 이름 변경
new.columns = ['연령대', '세종시인구수']
# 연령대구분 열 생성
new['연령대구분'] = ['총거주자수', '아동청소년', '청년', '중장년', '노년']
new
```

```python
# 동 개수만큼 세종특별자치시 인구수 추가 (21 * 5 = 105)
new2 = new
for i in range(20):
    new = new.append(new2)
new
```

```python
# 동 개수 list 생성
dong_list = f['읍면동'].unique()
dong_list
```

```python
# 동 list도 데이터프레임 행 개수 (21*5=105) 로 만들어줌
new_dong_list = []
for i in dong_list:
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
    new_dong_list.append(i)
new_dong_list
```

```python
# 데이터프레임에 읍면동 열 추가 후 동 이름 할당
new['읍면동'] = new_dong_list
new
```

```python
# 동 별 인구수 list 생성
population = []
for i in range(21):
    population.append(list(f.iloc[i][1:]))
population
```

```python
# 이제 인구수를 데이터프레임 개수만큼 할당하기 위해 인구수 2차원 list를 1차원 list로 변경 (5 * 21 = 105)
population2 = sum(population, [])
len(population2)
```

```python
# 동별인구수 열 생성
new['동별인구수'] = population2
new
```

```python
# 최종 데이터프레임 열 순서 변경
new = new[['읍면동', '동별인구수', '세종시인구수', '연령대', '연령대구분']]
new
```

```python
# 전처리 완료한 인구 현황 데이터 저장
new.to_csv('./data/202012인구_전처리완료.csv', index = False, encoding = 'utf-8')
```
