### 3번 파일 시군구(지역명) 열을 시군구, 동리, 읍면동으로 열 분리

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/3.세종시_아파트(매매)_실거래가.csv')
df
```

```python
df.columns = ['계약년월', '계약일', '지역명', '본번', '부번', '단지명', '전용면적(㎡)', '거래금액(만원)', '층',
       '건축년도', '해제사유발생일']
df
```

```python
lst = df['지역명'].str.split(' ')
lst_len = lst.str.len()
df['lst_len'] = lst_len
```

```python
df['lst_len'].value_counts()
```

```python
df
```

```python
df1 = df[df['lst_len'] == 2]
df1
```

```python
df1['시군구'] = df1['지역명'].str.split(' ').str[0]
df1['동리'] = df1['지역명'].str.split(' ').str[1]
df1
```

```python
df2 = df[df['lst_len'] == 3]
df2
```

```python
df2['시군구'] = df2['지역명'].str.split(' ').str[0]
df2['읍면동'] = df2['지역명'].str.split(' ').str[1]
df2['동리'] = df2['지역명'].str.split(' ').str[2]
df2
```

```python
del df1['lst_len']
del df2['lst_len']
```

```python
final_df = pd.concat([df1, df2])
final_df
```

```python
final_df.to_csv('./data/3.아파트_split.csv', index = False, encoding = 'utf-8')
```
